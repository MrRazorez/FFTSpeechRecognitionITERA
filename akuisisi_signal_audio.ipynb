```python
import soundfile as sf
import resampy
```

```python
file_path = str(input("Masukkan path\n"))  # Input path file audio yang sesuai
data, sample_rate = sf.read(file_path)
```

```python
# Mengubah sample rate menjadi 11.025 Hz
dst_sample_rate = 11025
resampled_data = resampy.resample(data, sample_rate, dst_sample_rate)
```

```python
import json

# Buat dictionary dengan data yang ingin Anda sertakan dalam JSON
json_data = {
    "resampled_data": resampled_data.tolist(),
    "dst_sample_rate": dst_sample_rate
}

# Ubah dictionary menjadi string JSON
json_string = json.dumps(json_data)

# Gunakan json_string untuk melakukan tindakan yang Anda inginkan, misalnya menyimpannya ke file
with open("akuisisi.json", "w") as json_file:
    json_file.write(json_string)
```

```python
# Melakukan plot sinyal audio
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 4))
plt.plot(resampled_data)
plt.title('Sonic Wave')
plt.xlabel('Waktu (detik)')
plt.ylabel('Amplitudo')
plt.show()
```
