```python
import json

# Baca file JSON dan baca isi JSON sebagai string
with open("akuisisi.json", "r") as json_file:
    json_string = json_file.read()

# Mengurai string JSON menjadi objek Python
json_data = json.loads(json_string)
```

```python
# Mengakses data dalam variabel json_data
resampled_data = json_data["resampled_data"]
dst_sample_rate = json_data["dst_sample_rate"]
```

```python
segment_duration = 0.2  # Durasi segmen dalam detik
segment_length = int(segment_duration * dst_sample_rate)  # Panjang segmen dalam jumlah sampel

segments = []
for i in range(0, len(resampled_data), segment_length):
    segment = resampled_data[i:i+segment_length]
    segments.append(segment)
```

```python
import json

# Buat dictionary dengan data yang ingin Anda sertakan dalam JSON
json_data = {
    "dst_sample_rate": dst_sample_rate,
    "segment_duration": segment_duration,
    "segment_length": segment_length,
    "segments": segments
}

# Ubah dictionary menjadi string JSON
json_string = json.dumps(json_data)

# Gunakan json_string untuk melakukan tindakan yang Anda inginkan, misalnya menyimpannya ke file
with open("ekstraksi.json", "w") as json_file:
    json_file.write(json_string)
```

```python
import matplotlib.pyplot as plt

for i, segment in enumerate(segments):
    plt.figure()
    plt.plot(segment)
    plt.title(f"Segment {i+1}")
    plt.xlabel("Sample")
    plt.ylabel("Amplitude")
    plt.show()
```
